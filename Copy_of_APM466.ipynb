```python
import numpy as np
import pandas as pd
```

## Template Bootstrap Function

```python
yields = pd.read_csv("yield.csv")
forwards = pd.read_csv("forward.csv")
yields
```

## Example Bond

```python
forwards
```

## Bootstrapping spot rates

```python
cov_y = np.cov(yields)
cov_f = np.cov(forwards)

cov_y
```

## Plotting the yield curve

```python
cov_f
```

```python
eigenval_y, eigenvec_y = np.linalg.eig(cov_y)
eigenval_y
#eigenvec_y
```

```python
eigenval_f, eigenvec_f = np.linalg.eig(cov_f)
eigenval_f
#eigenvec_f
```
